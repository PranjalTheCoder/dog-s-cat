```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
dataset_url = "https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip"
dataset_path = tf.keras.utils.get_file('cats_and_dogs.zip', origin=dataset_url, extract=True)
```

```python
import os
base_dir = os.path.join(os.path.dirname(dataset_path), 'cats_and_dogs_filtered')

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
```

```python
image_size = (150, 150)
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1.0/255,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1.0/255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=image_size,
                                                    batch_size=batch_size,
                                                    class_mode='binary')

validation_generator = val_datagen.flow_from_directory(validation_dir,
                                                       target_size=image_size,
                                                       batch_size=batch_size,
                                                       class_mode='binary')
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Binary classification
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(train_generator,
                    validation_data=validation_generator,
                    steps_per_epoch=100,
                    epochs=10,
                    validation_steps=50)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(10)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

```python
# Save the model
model.save('cats_and_dogs_classifier.h5')

# Load the model (if needed later)
from tensorflow.keras.models import load_model
model = load_model('cats_and_dogs_classifier.h5')

```

```python
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np

# Load and preprocess the image
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(150, 150))  # Resize the image
    img_array = img_to_array(img)                      # Convert to array
    img_array = np.expand_dims(img_array, axis=0)      # Add batch dimension
    img_array /= 255.0                                 # Normalize to [0, 1]
    return img_array

# Path to your new image
new_image_path = 'cat.webp'

# Preprocess the image
processed_image = preprocess_image(new_image_path)
```

```python
# Predict using the model
prediction = model.predict(processed_image)

# Interpret the result
if prediction[0] > 0.5:
    print("The image is a Dog.")
else:
    print("The image is a Cat.")
```

```python
import matplotlib.pyplot as plt

# Load the image again for display
image = load_img(new_image_path, target_size=(150, 150))

plt.imshow(image)
plt.axis('off')

# Display the prediction
label = "Dog" if prediction[0] > 0.5 else "Cat"
plt.title(f"Prediction: {label}")
plt.show()
```
